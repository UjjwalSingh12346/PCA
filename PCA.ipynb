```python
#Task 2: Dimensionality Reduction with PCA
#Standardize the features to ensure they have a mean of 0 and a standard deviation of 1:
from sklearn.preprocessing import StandardScaler

# Standardize features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
scaled_data
```

```python
#Implement PCA to reduce the dimensionality of the dataset:

from sklearn.decomposition import PCA
```

```python
# Apply PCA
pca = PCA()
pca_data = pca.fit_transform(scaled_data)

# Explained variance ratio
print(pca.explained_variance_ratio_)
```

```python
#Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance:

# Plotting explained variance
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

```python
# Cumulative explained variance
cumulative_variance = pca.explained_variance_ratio_.cumsum()
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.show()
```

```python
#Transform the original dataset into the principal components:
# Transform the data
pca_data = pca.transform(scaled_data)
```

```python
pca_data
```

```python
#Task 3: Clustering with Original Data
#Apply a clustering algorithm (e.g., K-means) to the original dataset:
from sklearn.cluster import KMeans

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(scaled_data)
clusters_original = kmeans.labels_
```

```python
clusters_original
```

```python
#Visualize the clustering results using appropriate plots:

plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=clusters_original, cmap='viridis')
plt.title('KMeans Clustering (Original Data)')
plt.show()
```

```python
#Evaluate the clustering performance using metrics such as silhouette score or Daviesâ€“Bouldin index:
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Evaluate clustering performance
silhouette = silhouette_score(scaled_data, clusters_original)
davies_bouldin = davies_bouldin_score(scaled_data, clusters_original)

print(f'Silhouette Score: {silhouette}')
print(f'Davies-Bouldin Score: {davies_bouldin}')
```

```python
#Task 4: Clustering with PCA Data
kmeans_pca = KMeans(n_clusters=3)
kmeans_pca.fit(pca_data)
clusters_pca = kmeans_pca.labels_
```

```python
clusters_pca
```

```python
#Visualize the clustering results obtained from PCA-transformed data
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=clusters_pca, cmap='viridis')
plt.title('KMeans Clustering (PCA Data)')
plt.show()
```

```python
#Compare the clustering results from PCA-transformed data with those from the original dataset:
# Compare silhouette scores
silhouette_pca = silhouette_score(pca_data, clusters_pca)
print(f'Silhouette Score (PCA): {silhouette_pca}')
```
